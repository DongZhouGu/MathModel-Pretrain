```python
best_X_df = pd.read_csv("迭代最优个体_20_2.csv",index_col=0)
best_X_df
```

```python
best_X_df.iloc[5:,:]
```

```python
best_X_feature = best_X_df.iloc[5:,:-6]
best_X_feature.max(axis=0) - best_X_feature.min(axis=0)
```

```python
best_X_feature.loc[:,(best_X_feature.max(axis=0) - best_X_feature.min(axis=0)) != 0]
```

```python
best_X_feature.loc[:,best_X_feature.std() >= 1]
```

```python
std_values =  np.sort(best_X_feature.std().values)
df = pd.DataFrame(columns=['阈值','特征数量'],index=np.arange(0,10,1))

for thres in np.arange(0,10,1):
    df.iloc[thres,0] = thres
    df.iloc[thres,1] = (std_values >= thres).sum()
df
```

```python
import matplotlib.pyplot as plt
sns.set(style="white")
# 设置中文编码和负号的正常显示
plt.rcParams['font.sans-serif'] = 'SimHei'
 
plt.tight_layout()
sns.barplot(x=df['阈值'],y=df['特征数量'],color='dodgerblue')
plt.title('方差分布图')
plt.xlabel('方差阈值')
plt.ylabel('特征数目')
plt.savefig("方差分布图.png",dpi=400)
```

```python
aa = best_X_feature.loc[:,best_X_feature.std() >= 1]
aa
```
